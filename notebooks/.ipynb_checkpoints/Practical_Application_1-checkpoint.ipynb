```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style for the project
sns.set_theme(style="whitegrid")

# Load the dataset
# Ensure coupons.csv is in a folder named 'data' one level up, or adjust path as needed
df = pd.read_csv('../data/coupons.csv')

print(f"Dataset loaded with {df.shape[0]} rows and {df.shape[1]} columns.")
df.head()
```

```python
# 1. Inspect missing values
print("Missing values per column:\n", df.isnull().sum())

# 2. Drop the 'car' column (too many missing values to be useful)
df_clean = df.drop(columns=['car'])

# 3. Handle missing values in frequency columns
freq_cols = ['Bar', 'CoffeeHouse', 'CarryAway', 'RestaurantLessThan20', 'Restaurant20To50']
df_clean[freq_cols] = df_clean[freq_cols].fillna('never')

# 4. Correct column typo
df_clean = df_clean.rename(columns={'passanger': 'passenger'})

print("\nData cleaning complete. Key analytical columns have been prepared for analysis.")
```

## Data Cleaning

Before beginning the analysis, I inspected the dataset for missing values. The `car` column was dropped because it contained too many missing values to be reliable for analysis. Missing values in the behavioral frequency columns (`Bar`, `CoffeeHouse`, `CarryAway`, `RestaurantLessThan20`, and `Restaurant20To50`) were filled with `"never"` under the assumption that a missing response indicates no regular visits. I also corrected the misspelled `passanger` column to `passenger` for consistency.

```python
# Overall acceptance rate
overall_rate = df_clean['Y'].mean()
print(f"Overall Acceptance Rate (Y=1): {overall_rate:.2%}")

# Summary by Coupon Type
coupon_summary = df_clean.groupby('coupon')['Y'].value_counts(normalize=True).unstack()
print("\nAcceptance Proportion by Coupon Category:")
print(coupon_summary)

# Summary by Passenger Type
passenger_summary = df_clean.groupby('passenger')['Y'].mean().sort_values(ascending=False)
print("\nAcceptance Rate by Passenger Type:")
print(passenger_summary)
```

```python
# Additional summary: acceptance by age
age_summary = df_clean.groupby('age')['Y'].mean().sort_values(ascending=False)
print("Acceptance Rate by Age:")
print(age_summary)
```

## Initial Observations

The overall coupon acceptance rate gives a baseline understanding of how often customers accept offers. Breaking the data down by coupon type shows that some coupons are more appealing than others. Acceptance rates by passenger type suggest that social context may influence a driver's decision to use a coupon. These early summaries indicate that both the type of coupon and the surrounding driving situation matter.

```python
# Plot 1: Acceptance Rate by Coupon Category
plt.figure(figsize=(12, 6))
order = df_clean.groupby('coupon')['Y'].mean().sort_values(ascending=False).index
sns.barplot(
    data=df_clean,
    x='coupon',
    y='Y',
    hue='coupon',
    palette='viridis',
    order=order,
    legend=False,
    errorbar=None
)
plt.title('Coupon Acceptance Rate by Category')
plt.xlabel('Coupon Type')
plt.ylabel('Acceptance Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../images/acceptance_by_type.png')
plt.show()

# Plot 2: Coffee House Acceptance Rate by Visit Frequency
plt.figure(figsize=(10, 6))
coffee_df = df_clean[df_clean['coupon'] == 'Coffee House']
freq_order = ['never', 'less1', '1~3', '4~8', 'gt8']
sns.barplot(
    data=coffee_df,
    x='CoffeeHouse',
    y='Y',
    hue='CoffeeHouse',
    order=freq_order,
    palette='magma',
    legend=False,
    errorbar=None
)
plt.title('Coffee House Coupon Acceptance by Visit Frequency')
plt.xlabel('Coffee House Visit Frequency')
plt.ylabel('Acceptance Rate')
plt.tight_layout()
plt.savefig('../images/coffeehouse_frequency.png')
plt.show()
```

## Visualization Insights

The first chart shows that coupon acceptance varies noticeably by coupon category, which suggests that some offers are naturally more attractive to customers. The second chart shows a relationship between coffee house visit frequency and acceptance of coffee house coupons. Customers who visit coffee houses more often appear more likely to accept these coupons, suggesting that prior customer behavior is an important predictor of acceptance.

```python
# Analyze Bar coupon acceptance
bar_coupons = df_clean[df_clean['coupon'] == 'Bar']

print("Bar coupon acceptance rate:")
print(bar_coupons['Y'].mean())

print("\nAcceptance by bar visit frequency:")
print(bar_coupons.groupby('Bar')['Y'].mean().sort_index())

print("\nAcceptance by age:")
print(bar_coupons.groupby('age')['Y'].mean().sort_values(ascending=False))

print("\nAcceptance by passenger:")
print(bar_coupons.groupby('passenger')['Y'].mean().sort_values(ascending=False))
```

```python
plt.figure(figsize=(10, 6))
bar_order = ['never', 'less1', '1~3', '4~8', 'gt8']
sns.barplot(
    data=bar_coupons,
    x='Bar',
    y='Y',
    hue='Bar',
    order=bar_order,
    palette='coolwarm',
    legend=False,
    errorbar=None
)
plt.title('Bar Coupon Acceptance by Bar Visit Frequency')
plt.xlabel('Bar Visit Frequency')
plt.ylabel('Acceptance Rate')
plt.tight_layout()
plt.savefig('../images/bar_frequency.png')
plt.show()
```

## Deeper Analysis: Bar Coupons

To better understand customer behavior, I focused on bar coupons. The results suggest that customers who go to bars more frequently are more likely to accept bar-related coupons. This supports the idea that existing habits strongly influence coupon acceptance. Differences across age groups and passenger types may also help identify which customer segments are more responsive to this type of promotion.

## Conclusions

This analysis shows that coupon acceptance depends on both customer behavior and driving context. Acceptance rates vary by coupon type, indicating that some offers are more appealing than others. Customers are also more likely to accept coupons connected to places they already visit frequently, such as coffee houses or bars. In addition, factors such as passenger type and age appear to influence acceptance patterns.

Overall, coupon acceptance is not random. Behavioral history and situational context both play an important role in whether a customer accepts an offer.

## Recommendations

Based on this analysis, companies should consider targeting customers with coupons that match their existing habits. For example, customers who frequently visit coffee houses or bars may be more likely to respond to coupons in those categories. Businesses should also consider situational factors such as who the customer is traveling with and the type of destination involved. Instead of distributing all coupons equally, marketers should prioritize the categories and customer segments with the highest historical acceptance rates.
